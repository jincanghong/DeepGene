## 将snp序列编码，并缓存

```python
data=pd.read_csv('core/core.tab',sep='\t',encoding='utf-8')
data
```

## 直接进行编码

```python
pos=data['POS'].values
data1=data[data.columns[3:]]
data1
```

```python
data1[data1=='A']=2
data1[data1=='T']=2
data1[data1=='C']=2
data1[data1=='G']=3
data1
```

```python
data2=data
for name in data2.columns[3:]:
    data2[name][data2[name]==data2['REF']]='N'
data2
```

```python
data2=data2[data2.columns[3:]]
data2[data2=='A']=0
data2[data2=='T']=1
data2[data2=='C']=2
data2[data2=='G']=3
data2[data2=='N']=4
```

```python
data3=data
seq_base = {}
for name in data3.columns[3:]:
    data3[name] = data3['REF']+data3[name]
    new_keys = data3[name].value_counts().keys()
    for k in new_keys:
        if k not in seq_base: seq_base[k]=len(seq_base)
data3
```
